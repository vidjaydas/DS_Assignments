```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
import re
import string
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
text = open("/content/Elon_musk.csv", encoding='cp1252')
  
# joining with space content of text
text = ' '.join([i for i in text])  
  
# replacing ',' by space
text = text.replace(",", " ")  
text = re.sub(r'\w+:\/{2}[\d\w-]+(\.[\d\w-]+)*(?:(?:\/[^\s/]*))*', '', text)
  
#displaying result
print(text)
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
wordnet = WordNetLemmatizer
nltk.download('wordnet')
```

```python
tweet_rev_string = ''.join(text)
```

```python
tweet_rev_string
```

```python
tweet_rev_string = re.sub("[^A-Za-z" "]+"," ",tweet_rev_string).lower()
tweet_rev_string =re.sub("[0-9" "]+"," ",tweet_rev_string)
```

```python
tweet_rev_words = tweet_rev_string.split(" ")
```

```python
tweet_rev_words = [w for w in tweet_rev_words if not w in stop_words]
```

```python
tweet_rev_string =" ".join(tweet_rev_words)
```

```python
wordcloud_ip = WordCloud(
                      background_color='black',
                      width=1800,
                      height=1400
                     ).generate(tweet_rev_string)
```
